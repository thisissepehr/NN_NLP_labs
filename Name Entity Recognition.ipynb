**Taks1 : explaination**

To build the model we need to put two GRU elements in Bidirectional elements. We need the state sequences after processing to pass them to FFNNs so we set the ``` return_sequences ``` to true. Also, to avoid overfitting dropout rate of 0.2 is added to recurrent GRUs. <br>
After that there is a loop which creates the ```Dense ``` layers followed by a 0.2 ```Dropout``` layer. finally the final classifier is set with 5 outputs which are the tags of NER. Since we have multiple labels to classify, the activation function for the last layer is softmax.



**Task2: explaination**<br>
As the following graph shows the model architecture is implemented exactly like the presented paper. By training this model we can reach 81.5% for f1 score on test set. The model runs smoothly in GPU with 6 minutes spent on trining.The training process log is available in the above block. <br>
The evaluation function retrieves the predictions and compares them with gold labels and if there is a mismatch between them wheather its in the index of the label it self it is not considered as True positive(strict f1 score).
